```python
# now split the data
# import the sklearn library with train, test,split function

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
# created algo
reg.fit(x_train, y_train)
```

```python
hd.head()
```

```python
# now preducted price noted as y_pred

y_pred=reg.predict(x_test)
y_pred
```

```python
# visualization of training phase
plt.scatter(x_train, y_train, color="orange")
plt.plot(x_train, reg.predict(x_train), color='red')
plt.title('price vs space')
plt.xlabel('space')
plt.ylabel('price')
plt.show()
```

```python
# visaulization with testing phase
plt.scatter(x_test, y_test, color='yellow')
plt.plot(x_train, reg.predict(x_train), color='green')
plt.title('space vs price')
plt.xlabel('space')
plt.ylabel('price')
plt.show()
```
